```python
class MultiTapeTuringMachine:
    def __init__(self, num_tapes, initial_state, final_states):
        self.num_tapes = num_tapes
        self.tapes = [""] * num_tapes
        self.heads = [0] * num_tapes
        self.state = initial_state
        self.final_states = final_states

    # Funcion para tomar las cabezas de las cintas:
    def get_heads(self):
        return self.heads
    
    # Funcion para tomar las cintas de la MT:
    def get_tapes(self):
        return self.tapes
    
    # Funcion para el indice de la cinta en la que se encuentra la MT:
    def get_tape_index(self, tape):
        for i in range(self.tapes):
            if tape == self.tapes[i]:
                return i
    
    # Funcion para el indice de la cabeza en la que se encuentra la MT:
    def get_head_index(self, head):
        for i in range(self.heads):
            if head == self.heads[i]:
                return i

    # Funcion para tomar el simbolo de la cinta en la que se encuentra la MT:
    def get_symbol(self, tape_index):
        return self.tapes[tape_index][self.heads[tape_index]]
    
    
    # Funcion para tomar la cinta en la que se encuentra la MT:
    def get_tape(self, tape_index):
        return self.tapes[tape_index]
    
    # Funcion para tomar la cabeza de la cinta en la que se encuentra la MT:
    def get_head(self, head_index):
        return self.heads[head_index]

    # Funcion para mover la cabeza de la cinta a la derecha:
    def moveright(self, tape_index, letter):
        self.set_symbol(tape_index, letter)
        self.heads[tape_index] += 1

    # Funcion para mover la cabeza de la cinta a la izquierda:
    def moveleft(self, tape_index, letter):
        self.set_symbol(tape_index, letter)
        self.heads[tape_index] -= 1

    def write(self, tape_index):
        self.tapes[tape_index].append("Ñ")


    # Funcion para mover la cabeza de la cinta:
    def move(self, tape_index, direction, letter):
        if direction == 'R':
            self.moveright(tape_index, letter)
        elif direction == 'L':
            self.moveleft(tape_index, letter)

        self.write(tape_index)

    # Funcion para incialzar el simbolo de la cinta que se desee en la MT:
    def set_symbol(self, tape_index, symbol):
        self.tapes[tape_index][self.heads[tape_index]] = symbol

    # Funcion para incialzar el estado de la cinta que se desee en la MT:
    def set_state(self, state):
        self.state = state
    
    # Funcion para inicializar o crear las cintas de la MT:
    def initialize_tapes(self, inputs):
        for i in range(self.num_tapes):
            self.tapes[i] = list("ÑÑ" + inputs[i] + "ÑÑ")
            self.heads[i] = 0

        for i in range(self.num_tapes):
            print(self.tapes[i])

    def step_q0(self, symbols):
        if symbols[0] != symbols[2]:
            self.set_state("q0")
            self.move(2, "R", symbols[2])
            self.move(3, "R", 1)

        elif symbols[0] == symbols[2]:
            if symbols[0] != 'Ñ':
                self.set_state("q1")
                self.move(0, "R", symbols[0])
                self.move(2, "L", symbols[2])
            else:
                self.move(0, "R", symbols[0])
                self.move(1, "R", symbols[1])
                self.move(2, "R", symbols[2])
                self.move(3, "R", symbols[3])
                self.move(4, "R", symbols[4])
        
    
    def step_q1(self, symbols):
        if (symbols[0] != "Ñ") & (symbols[2]!="Ñ"):
            self.set_state("q1")
            self.move(2, "L", symbols[2])

        elif (symbols[0] != "Ñ") & (symbols[2]=="Ñ"):
            self.set_state("q2")
            self.move(2, "R", symbols[2])

        elif (symbols[0] == "Ñ") & (symbols[2]!="Ñ"):
            self.set_state("q1")
            self.move(2, "L", symbols[2])

        elif (symbols[0] == "Ñ") & (symbols[2]=="Ñ"):
            self.set_state("q2")
            self.move(2, "R", symbols[2])
    
    def step_q2(self, symbols):
        if (symbols[0] != "Ñ") & (symbols[1]!=symbols[2]):
            self.set_state("q2")
            self.move(2, "R", symbols[2])
            self.move(3, "R", 1)

        elif (symbols[0] != "Ñ") & (symbols[1]== symbols[2]):
            self.set_state("q3")
            self.move(1, "R", symbols[1])
            self.move(2, "L", symbols[2])
            self.move(3, "R", "Ñ")

        elif (symbols[0] == "Ñ") & (symbols[1]!= symbols[2]):
            self.set_state("q2")
            self.move(2, "R", symbols[2])
            self.move(3, "R", 1)

        elif (symbols[0] == "Ñ") & (symbols[1]== symbols[2]):
            self.set_state("q3")
            self.move(2, "L", symbols[2])
        
    def step_q3(self, symbols):
        print("Estoy dentro de q3")
        if (symbols[0] != "Ñ") & (symbols[1]!= "Ñ" ) & (symbols[2] != "Ñ"):
            print("1 if")
            self.set_state("q3")
            self.move(2, "L", symbols[2])

        elif (symbols[0] != "Ñ") & (symbols[1]!= "Ñ" ) & (symbols[2] == "Ñ"):
            print("2 if")
            self.set_state("q0")
            self.move(2, "R", symbols[2])

        elif (symbols[0] != "Ñ") & (symbols[1]== "Ñ" ) & (symbols[2] != "Ñ"):
            print("3 if")
            self.set_state("q4")
            self.move(1, "L", symbols[1])
            self.move(2, "L", symbols[2])

        elif (symbols[0] == "Ñ") & (symbols[1]!= "Ñ" ) & (symbols[2] != "Ñ"):
            print("4 if")
            self.set_state("q3")
            self.move(2, "L", symbols[2])
        
        elif (symbols[0] == "Ñ") & (symbols[1]!= "Ñ" ) & (symbols[2] == "Ñ"):
            print("5 if")
            self.set_state("q5")
            self.move(2, "R", symbols[2])
            self.move(3, "L", symbols[3])
    
    def step_q4(self, symbols):
        if (symbols[1] != "Ñ"):
            self.set_state("q4")
            self.move(1, "L", symbols[1])

        elif (symbols[1] == "Ñ"):
            self.set_state("q3")
            self.move(1, "R", symbols[1])

    def step_q5(self, symbols):
        if (symbols[3] != "Ñ"):
            self.set_state("q5")
            self.move(3, "L", symbols[3])

        elif (symbols[3] == "Ñ"):
            self.set_state("q6")
            self.move(3, "L", symbols[3])
    
    def step_q6(self, symbols):
        if (symbols[3] != "Ñ"):
            self.set_state("q5")
            self.move(3, "L", symbols[3])

        elif (symbols[3] == "Ñ"):
            self.set_state("q7")
            self.move(3, "R", symbols[3])
    
    def step_q7(self, symbols):
        if (symbols[0] == "Ñ"):
            self.set_state("q8")
            self.move(3, "R", symbols[3])

    def step_q8(self, symbols):
        if (symbols[3] != "Ñ"):
            self.set_state("q8")
            self.move(2, "R", symbols[2])
            self.move(3, "R", symbols[3])

        elif (symbols[3] == "Ñ"):
            self.set_state("q9")
            self.move(2, "L", symbols[2])
            self.move(3, "R", symbols[3])
            
    def step_q9(self, symbols):
        if (symbols[3] != "Ñ"):
            self.set_state("q8")
            self.move(4, "R", symbols[2])
            

        elif (symbols[3] == "Ñ"):
            self.set_state("q10")
            self.move(0, "L", symbols[0])
            self.move(4, "L", symbols[4])
            
    def step_q10(self, symbols):
        if (symbols[0] != "Ñ") & (symbols[4] != "Ñ"):
            self.set_state("q10")
            self.move(0, "L", "Ñ")
            self.move(4, "L", symbols[4])

        elif (symbols[0] == "Ñ") & (symbols[4] == "Ñ"):
            self.set_state("q11")
            self.move(0, "R", symbols[0])
            self.move(4, "R", symbols[4])
            
    def step_q11(self, symbols):
        if (symbols[4] != "Ñ"):
            self.set_state("q11")
            self.move(0, "R", symbols[4])
            self.move(4, "R", symbols[4])

        elif (symbols[4] == "Ñ"):
            self.set_state("qf")
            

    
    def step(self, symbols, state):
        if state == "q0":
            self.step_q0(symbols)
            
        elif state == "q1":
            self.step_q1(symbols)
            
        elif state == "q2":
            self.step_q2(symbols)
            
        elif state == "q3":
            print(symbols)
            self.step_q3(symbols)
            
        elif state == "q4":
            self.step_q4(symbols)
            
        elif state == "q5":
            self.step_q5(symbols)
            
        elif state == "q6":
            self.step_q6(symbols)
            
        elif state == "q7":
            self.step_q7(symbols)
            
        elif state == "q8":
            self.step_q8(symbols)
            
        elif state == "q9":
            self.step_q9(symbols)
            
        elif state == "q10":
            self.step_q10(symbols)
            
        elif state == "q11":
            self.step_q11(symbols)
            
        elif state == "qf":
            
            return True
        else:
            return False
        return True
            

    # Funcion de transición de la MT:
    def transition(self):
        symbols = []
        for i in range(0,5):
            symbols.append(self.get_symbol(i))
        
        return self.step(symbols, self.state)


    def run(self):
        while self.state not in self.final_states:
            if not self.transition():
                print("No hay transición definida para:", self.state)
                break
            # self.print_tapes()


    def print_tapes(self):
        print("Estado actual:", self.state)
        for i in range(self.num_tapes):
            tape_str = ''.join(self.tapes[i])
            head_pos = self.heads[i]
            print(f"Cinta {i+1}: {tape_str}")
            print("          " + " " * head_pos + "^")
        print("----------")

```
