```python
class MultiTapeTuringMachine:
    def __init__(self, num_tapes, initial_state, final_states, f_transition):
        self.num_tapes = num_tapes
        self.tapes = [[]] * num_tapes
        self.heads = [0] * num_tapes
        self.state = initial_state
        self.final_states = final_states
        self.f_transition = f_transition

    # Funcion para tomar las cabezas de las cintas:
    def get_heads(self):
        return self.heads
    
    # Funcion para tomar las cintas de la MT:
    def get_tapes(self):
        return self.tapes
    
    # Funcion para el indice de la cinta en la que se encuentra la MT:
    def get_tape_index(self, tape):
        for i in range(self.tapes):
            if tape == self.tapes[i]:
                return i
    
    # Funcion para el indice de la cabeza en la que se encuentra la MT:
    def get_head_index(self, head):
        for i in range(self.heads):
            if head == self.heads[i]:
                return i

    # Funcion para tomar el simbolo de la cinta en la que se encuentra la MT:
    def get_symbol(self, tape_index):
        return self.tapes[tape_index][self.heads[tape_index]]
    
    
    # Funcion para tomar la cinta en la que se encuentra la MT:
    def get_tape(self, tape_index):
        return self.tapes[tape_index]
    
    # Funcion para tomar la cabeza de la cinta en la que se encuentra la MT:
    def get_head(self, head_index):
        return self.heads[head_index]

    # Funcion para mover la cabeza de la cinta a la derecha:
    def moveright(self, tape_index, letter):
        self.set_symbol(tape_index, letter)
        self.heads[tape_index] += 1

    # Funcion para mover la cabeza de la cinta a la izquierda:
    def moveleft(self, tape_index, letter):
        self.tapes[tape_index][self.heads[tape_index]+1] = letter
        self.heads[tape_index] -= 1

    # Funcion para mover la cabeza de la cinta:
    def move(self, tape_index, direction, letter):
        if direction == 'R':
            self.moveright(tape_index, letter)
        elif direction == 'L':
            self.moveleft(tape_index, letter)

    # Funcion para incialzar el simbolo de la cinta que se desee en la MT:
    def set_symbol(self, tape_index, symbol):
        self.tapes[tape_index][self.heads[tape_index]] = symbol

    # Funcion para incialzar el estado de la cinta que se desee en la MT:
    def set_state(self, state):
        self.state = state
    
    # Funcion para inicializar o crear las cintas de la MT:
    def initialize_tapes(self, inputs):
        for i in range(len(self.num_tapes)):
            self.tapes[i] = ["B"] + ["B"] + inputs[i] + ["B"] + ["B"]
            self.heads[i] = 0

    def step_q0(self, symbols):
        if symbols[0] != symbols[2]:
            self.set_state(0, "q0")
            self.set_state(1, "q0")
            self.set_state(2, "q0")
            self.set_state(3, "q0")
            self.set_state(4, "q0")
            self.move(2, "R", symbols[2])
            self.move(3, "R", 1)

        elif symbols[0] == symbols[2]:
            self.set_state(0, "q1")
            self.set_state(1, "q1")
            self.set_state(2, "q1")
            self.set_state(3, "q1")
            self.set_state(4, "q1")
            self.move(0, "R", symbols[0])
            self.move(2, "L", symbols[2])
    
    def step_q1(self, symbols):
        if (symbols[0] != ["B"]) & (symbols[2]!=["B"]):
            self.set_state(0, "q1")
            self.set_state(1, "q1")
            self.set_state(2, "q1")
            self.set_state(3, "q1")
            self.set_state(4, "q1")
            self.move(2, "L", symbols[2])

        elif (symbols[0] != ["B"]) & (symbols[2]==["B"]):
            self.set_state(0, "q2")
            self.set_state(1, "q2")
            self.set_state(2, "q2")
            self.set_state(3, "q2")
            self.set_state(4, "q2")
            self.move(2, "R", symbols[2])

        elif (symbols[0] == ["B"]) & (symbols[2]!=["B"]):
            self.set_state(0, "q1")
            self.set_state(1, "q1")
            self.set_state(2, "q1")
            self.set_state(3, "q1")
            self.set_state(4, "q1")
            self.move(2, "L", symbols[2])

        elif (symbols[0] == ["B"]) & (symbols[2]==["B"]):
            self.set_state(0, "q2")
            self.set_state(1, "q2")
            self.set_state(2, "q2")
            self.set_state(3, "q2")
            self.set_state(4, "q2")
            self.move(2, "R", symbols[2])
    
    def step_q2(self, symbols):
        if (symbols[0] != ["B"]) & (symbols[1]!=symbols[2]):
            self.set_state(0, "q2")
            self.set_state(1, "q2")
            self.set_state(2, "q2")
            self.set_state(3, "q2")
            self.set_state(4, "q2")
            self.move(2, "R", symbols[2])
            self.move(3, "R", 1)

        elif (symbols[0] != ["B"]) & (symbols[1]== symbols[2]):
            self.set_state(0, "q3")
            self.set_state(1, "q3")
            self.set_state(2, "q3")
            self.set_state(3, "q3")
            self.set_state(4, "q3")
            self.move(1, "R", symbols[1])
            self.move(2, "L", symbols[2])
            self.move(3, "R", "B")

        elif (symbols[0] == ["B"]) & (symbols[1]!= symbols[2]):
            self.set_state(0, "q2")
            self.set_state(1, "q2")
            self.set_state(2, "q2")
            self.set_state(3, "q2")
            self.set_state(4, "q2")
            self.move(2, "R", symbols[2])
            self.move(3, "R", 1)

        elif (symbols[0] == ["B"]) & (symbols[1]== symbols[2]):
            self.set_state(0, "q3")
            self.set_state(1, "q3")
            self.set_state(2, "q3")
            self.set_state(3, "q3")
            self.set_state(4, "q3")
            self.move(2, "L", symbols[2])
        
    def step_q3(self, symbols):
        if (symbols[0] != ["B"]) & (symbols[1]!= ["B"] ) & (symbols[2] != ["B"]):
            self.set_state(0, "q3")
            self.set_state(1, "q3")
            self.set_state(2, "q3")
            self.set_state(3, "q3")
            self.set_state(4, "q3")
            self.move(2, "L", symbols[2])

        elif (symbols[0] != ["B"]) & (symbols[1]!= ["B"] ) & (symbols[2] == ["B"]):
            self.set_state(0, "q0")
            self.set_state(1, "q0")
            self.set_state(2, "q0")
            self.set_state(3, "q0")
            self.set_state(4, "q0")
            self.move(2, "R", symbols[2])

        elif (symbols[0] != ["B"]) & (symbols[1]== ["B"] ) & (symbols[2] != ["B"]):
            self.set_state(0, "q4")
            self.set_state(1, "q4")
            self.set_state(2, "q4")
            self.set_state(3, "q4")
            self.set_state(4, "q4")
            self.move(1, "L", symbols[1])
            self.move(2, "L", symbols[2])

        elif (symbols[0] == ["B"]) & (symbols[1]!= ["B"] ) & (symbols[2] != ["B"]):
            self.set_state(0, "q3")
            self.set_state(1, "q3")
            self.set_state(2, "q3")
            self.set_state(3, "q3")
            self.set_state(4, "q3")
            self.move(2, "L", symbols[2])
        
        elif (symbols[0] == ["B"]) & (symbols[1]!= ["B"] ) & (symbols[2] == ["B"]):
            self.set_state(0, "q5")
            self.set_state(1, "q5")
            self.set_state(2, "q5")
            self.set_state(3, "q5")
            self.set_state(4, "q5")
            self.move(2, "R", symbols[2])
            self.move(3, "L", symbols[3])
    
    def step_q4(self, symbols):
        if (symbols[1] != ["B"]):
            self.set_state(0, "q4")
            self.set_state(1, "q4")
            self.set_state(2, "q4")
            self.set_state(3, "q4")
            self.set_state(4, "q4")
            self.move(1, "L", symbols[1])

        elif (symbols[1] == ["B"]):
            self.set_state(0, "q3")
            self.set_state(1, "q3")
            self.set_state(2, "q3")
            self.set_state(3, "q3")
            self.set_state(4, "q3")
            self.move(1, "R", symbols[1])

    def step_q5(self, symbols):
        if (symbols[3] != ["B"]):
            self.set_state(0, "q5")
            self.set_state(1, "q5")
            self.set_state(2, "q5")
            self.set_state(3, "q5")
            self.set_state(4, "q5")
            self.move(3, "L", symbols[3])

        elif (symbols[3] == ["B"]):
            self.set_state(0, "q6")
            self.set_state(1, "q6")
            self.set_state(2, "q6")
            self.set_state(3, "q6")
            self.set_state(4, "q6")
            self.move(3, "L", symbols[3])
    
    def step_q6(self, symbols):
        if (symbols[3] != ["B"]):
            self.set_state(0, "q5")
            self.set_state(1, "q5")
            self.set_state(2, "q5")
            self.set_state(3, "q5")
            self.set_state(4, "q5")
            self.move(3, "L", symbols[3])

        elif (symbols[3] == ["B"]):
            self.set_state(0, "q7")
            self.set_state(1, "q7")
            self.set_state(2, "q7")
            self.set_state(3, "q7")
            self.set_state(4, "q7")
            self.move(3, "R", symbols[3])
    
    def step_q7(self, symbols):
        if (symbols[0] == ["B"]):
            self.set_state(0, "q8")
            self.set_state(1, "q8")
            self.set_state(2, "q8")
            self.set_state(3, "q8")
            self.set_state(4, "q8")
            self.move(3, "R", symbols[3])

    def step_q8(self, symbols):
        if (symbols[3] != ["B"]):
            self.set_state(0, "q8")
            self.set_state(1, "q8")
            self.set_state(2, "q8")
            self.set_state(3, "q8")
            self.set_state(4, "q8")
            self.move(2, "R", symbols[2])
            self.move(3, "R", symbols[3])

        elif (symbols[3] == ["B"]):
            self.set_state(0, "q9")
            self.set_state(1, "q9")
            self.set_state(2, "q9")
            self.set_state(3, "q9")
            self.set_state(4, "q9")
            self.move(2, "L", symbols[2])
            self.move(3, "R", symbols[3])

    
    def step(self, symbols, state):
        if state == "q0":
            self.step_q0(symbols)
        elif state == "q1":
            self.step_q1(symbols)
        elif state == "q2":
            self.step_q2(symbols)
        elif state == "q3":
            self.step_q3(symbols)
        elif state == "q4":
            self.step_q4(symbols)
        elif state == "q5":
            self.step_q5(symbols)
        elif state == "q6":
            self.step_q6(symbols)
        elif state == "q7":
            self.step_q7(symbols)
        elif state == "q8":
            self.step_q8(symbols)
        elif state == "q9":
            self.step_q9(symbols)
        elif state == "q10":
            self.step_q10(symbols)
        elif state == "q11":
            self.step_q11(symbols)
        return True
            

    # Funcion de transición de la MT:
    def transition(self):
        symbols = [[]]
        for i in range(len(self.num_tapes)):
            symbols[i] = self.get_symbol(i)
        
        if (self.step(symbols, self.state)):
            return True
        else:
            return False

    
    

    def run(self):
        while self.state not in self.final_states:
            if not self.step():
                print("No hay transición definida para:", self.state)
                break
            self.print_tapes()

    def print_tapes(self):
        print("Estado actual:", self.state)
        for i in range(self.num_tapes):
            tape_str = ''.join(self.tapes[i])
            head_pos = self.heads[i]
            print(f"Cinta {i+1}: {tape_str}")
            print("          " + " " * head_pos + "^")
        print("----------")
```
