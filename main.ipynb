```python
# Simulación optimizada del cifrado Vigenère con cintas sin estados separados

# Inicialización de las cintas
def initialize_tapes(mensaje, clave, abecedario="ABCDEFGHIJKLMNOPQRSTUVWXYZ"):
    reps = (len(mensaje) // len(clave)) + 1
    clave_expandida = (clave * reps)[:len(mensaje)]

    cinta0 = list("ÑÑÑ" + mensaje + "ÑÑÑ")
    cinta1 = list("ÑÑÑ" + clave_expandida + "ÑÑÑ")
    cinta2 = list("ÑÑÑ" + abecedario + "ÑÑÑ")
    cinta3 = []
    cinta4 = []
    return cinta0, cinta1, cinta2, cinta3, cinta4

# Convierte una letra a su representación unitaria
def letra_a_unitario(letra, abecedario):
    unidad = []
    for l in abecedario:
        if l == letra:
            break
        unidad.append("1")
    return unidad

# Suma dos representaciones unitarias
def suma_unitaria(unit1, unit2):
    return unit1 + unit2

# Convierte una suma unitaria a letra (con rollover)
def unitario_a_letra(unit_suma, abecedario):
    count = 0
    for simbolo in unit_suma:
        if simbolo == "1":
            count += 1
    while count >= len(abecedario):
        count -= len(abecedario)
    return abecedario[count]

# Algoritmo principal optimizado
def vigenere_sin_operadores(mensaje, clave):
    cinta0, cinta1, cinta2, cinta3, cinta4 = initialize_tapes(mensaje, clave)
    abecedario = [c for c in cinta2 if c != 'Ñ']

    idx = 3  # primer índice real de letras en cinta 0 y 1
    while cinta0[idx] != 'Ñ' and cinta1[idx] != 'Ñ':
        l1 = cinta0[idx]
        l2 = cinta1[idx]

        u1 = letra_a_unitario(l1, abecedario)
        u2 = letra_a_unitario(l2, abecedario)

        suma = suma_unitaria(u1, u2)

        cinta3.extend(suma + ['Ñ'])  # separación por 'Ñ'

        letra_resultado = unitario_a_letra(suma, abecedario)
        cinta4.append(letra_resultado)

        idx += 1

    return cinta0, cinta1, cinta2, cinta3, cinta4

```

```python
def cifrar_vigenere():
    mensaje_claro = input("Introduce la palabra a cifrar: ")
    palabra_clave = input("Introduce la palabra clave: ")

    cintas = vigenere_sin_operadores(mensaje_claro, palabra_clave)
    print(f"Cinta1: {cintas[0]}")
    print(f"Cinta2: {cintas[1]}")
    print(f"Cinta3: {cintas[2]}")
    print(f"Cinta4: {cintas[3]}")
    print(f"Cinta5/ Mensaje cifrado: {cintas[4]}")
```
